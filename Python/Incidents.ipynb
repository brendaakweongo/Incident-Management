```python
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(io = "Incidents.xlsx")

#displaying data
df
```

```python
#Dropping unnecessary columns

to_drop = ['Time incident was completed','Deadline','Progress','Solution and Action Plan (RCA) (Details)','Comments']

df.drop(to_drop,inplace = True, axis= 1)

df
```

```python
#Removing duplicates
df = df.drop_duplicates()
df
```

```python
#Total Incidents
total = len(df)
total
```

```python
#Escalations to CTO
escalations = len(df[df["Escalations To CTO"] == "Escalated to CTO"])
escalations
```

```python
#Unresolved Incidents
unresolvedIncidents = len(df[df["Status of Trouble Ticket"] == "Open"])
unresolvedIncidents
```

```python
#Resolved Incidents
resolvedIncidents = len(df[df["Status of Trouble Ticket"] == "Closed"])
resolvedIncidents
```

```python
#Percentage of unresolved incidents
unresolvedPercent = (unresolvedIncidents/total)*100
unresolvedPercent
```

```python
#Percentage of resolved incidents
resolvedPercent = (resolvedIncidents/total)*100
resolvedPercent
```

```python
#Number of incidents per status
df['Status'].value_counts()
```

```python
df['Status'].value_counts().plot(kind='bar', color='green', width=0.3)

plt.title('Number of incidents per status')
plt.xlabel('Status')
plt.ylabel('Incidents')

plt.grid(color='grey', linestyle='--',linewidth='0.3')
plt.show()
```

```python
#Number of incidents per facility
df['Facility'].value_counts()
```

```python
df['Facility'].value_counts().plot(kind='bar', color='green', width=0.3)

plt.title('Number of incidents per facility')
plt.xlabel('Facility')
plt.ylabel('Incidents')

plt.grid(color='grey',linestyle='--',linewidth='0.3')
plt.show()
```

```python
# Plot of number of resolved and unresolved incidents per facility
df.groupby(["Facility","Status of Trouble Ticket"]).size().plot(kind='bar', color='green', width=0.3)

plt.title('Number of resolved and unresolved incidents per facility')
plt.xlabel('Facility')
plt.ylabel('Incidents')

plt.grid(color='grey',linestyle='--',linewidth='0.3')
plt.show()
```
